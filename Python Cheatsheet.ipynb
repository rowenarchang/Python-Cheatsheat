# Python Parameters

```python
Parameters
    The input (variables) required in some functions
    Parameters are placed one after another, separated by commas, within ()
    
    # example
    def ready_for_school(backpack, pencil_case):
      if (backpack == 'full' and pencil_case == 'full'):
        print ("I'm ready for school!")
        
Functions
    Can be defined using def so that the same code doesn't have to be retyped
    Functions may return a value using the return keyword
    
    # example
    def my_function(x):
      return x + 1

Calling Functions
    Call a preexisting function by writing the name of it
    ex: doHomework()
        variable = dataset.function
        
Function Arguments
    Parameters in python are variables — placeholders for the actual values the function needs. \
    When the function is called, these values are passed in as arguments.

    # example
    def sales(grocery_store, item_on_sale, cost):
      print(grocery_store + " is selling " + item_on_sale + " for " + cost) 
    sales("The Farmer’s Market", "toothpaste", "$1")
    
    The arguments passed through the function .sales() are "The Farmer’s Market", "toothpaste", "$1"
    The paraments are: grocery_store, item_on_sale, cost

Function Keyword Arguments
    Keyword arguments = when function arguments are passed using their names
    Keyword arguments can be passed in any order
    
    # example
    def findvolume(length=1, width=1, depth=1):
      print("Length = " + str(length))
      print("Width = " + str(width))
      print("Depth = " + str(depth))
      return length * width * depth;

    findvolume(1, 2, 3)
    findvolume(length=5, depth=2, width=4)
    findvolume(2, depth=3, width=4)
    
Variable Scope
    Local variable = defined withing a function 
    Global variable = defined outside a function, is accessible inside the function
    
While Loop
    The while loop will not run n times, but until a defined condition is no longer met
    while is rarely used, except in cases where a user's input is required
```

# Lists

```python
Lists []
    Ordered collections of items
    Uses square brackets, separated by commas
    Can contain multiple data types
    
    # example
    list_of_lists = [['a', 1], ['b', 2]]
    
Adding Lists
    Lists can be added using + 
    
    # example
    items = ['cake', 'cookie', 'bread']
    total_items = items + ['biscuit', 'tart']
    print(total_items)
    # Result: ['cake', 'cookie', 'bread', 'biscuit', 'tart']
    
List Method .append()
    Use .append() to add values to the end of a list
    
    orders = ['daisies', 'periwinkle']
    orders.append('tulips')
    print(orders)
    # Result: ['daisies', 'periwinkle', 'tulips']
    
Split Function
    The function split will split a string on a character
    
    #example
    if mydate is the string 9-26-2017, then mydate.split('-') would return the list ['9', '26', '2017']

Aggregating Iterables Using zip()
    zip() aggregates corresponding elements in iterables
    
    # example
    owners_names = ['Jenny', 'Sam', 'Alexis']
    dogs_names = ['Elphonse', 'Dr. Doggy DDS', 'Carter']
    owners_dogs = zip(owners_names, dogs_names)
    print(list(owners_dogs))
    # Result: [('Jenny', 'Elphonse'), ('Sam', 'Dr.Doggy DDS'), ('Alexis', 'Carter')]
    
Selecting Items from Lists
    The first item is at index 0. 
    Use list[index] to access a list element by index
    
    # example
    To select items from index `0` to `3`: print(items[:4])
    To select items from index `2` to the last item, inclusive: print(items[2:])
    To select the last element: my_list[-1].
    To select the last three elements: my_list[-3:].
    To select everything except the last two elements: my_list[:-2].

List Method .count()
    .count() searches a list for the search item (argument)
    
    # example
    backpack = ['pencil', 'pen', 'notebook', 'textbook', 'pen', 'highlighter', 'pen']
    numPen = backpack.count('pen')
    print(numPen)
    # Output: 3
    
List Method len()
    len() finds the number of items in the list
    
    # example
    knapsack = [2, 4, 3, 7, 10]
    size = len(knapsack)
    print(size) 
    # Output: 5
    
List Method .sort()
    .sort() modifies list to be in ascending or alphabetical order
    ex: exampleList.sort() outputs [1, 2, 3, 4]
    
    sorted() Function
    returns a new, sorted list 
    ex: sortedList = sorted(unsortedList)
    
List Slicing
    A slice is a sublist of a list
    myList[START_NUMBER:END_NUMBER]
    
    # example
    tools = ['pen', 'hammer', 'lever']
    tools_slice = tools[1:3] # ['hammer', 'lever']
```

# List Comprehension

```python
List Comprehension
    Allows you to create new lists in one line
    
    # example
    temperatures = [-5, 29, 26, -7, 1, 18, 12, 31]
    temperatures_adjusted = [temp + 20 for temp in temperatures]
    # temperatures_adjusted is now [15, 49, 46, 13, 21, 38, 32, 51]

    This list comprehension names an element in temperatures as temp \
    and modifies it with temp + 20
    
    Use this for complicated mathematical operations like 
    temperatures_F = [(9.0/5.0)*temp + 32 for temp in temperatures]
    
Using .zip() 
    Used for going through two lists in one list comprehension
    
    # example
    x_values_midpoints = [(x1 + x2)/2.0 for (x1, x2) in zip(x_values_1, x_values_2)]
```

```python
Apply Function apply()
    Applies a function on every value in a column and/or row
    Defaults to all functions in a column
    To perform on a row, use axis=1
    
    # This function doubles the input value
    def double(x):
      return 2*x

    # Apply this function to double every value in a specified column
    df.column1 = df.column1.apply(double)

    # Lambda functions can also be supplied to `apply()`
    df.column2 = df.column2.apply(lambda x : 3*x)

    # Applying to a row requires it to be called on the entire DataFrame
    df['newColumn'] = df.apply(lambda row: 
      row['column1'] * 1.5 + row['column2'],
      axis=1)
    
Lambda Function Syntax
    lambda x: [OUTCOME IF TRUE] if [CONDITIONAL] else [OUTCOME IF FALSE]
    
Applying a Lambda to a Column
    # example
    
    get_last_name = lambda x: x.split()[-1]
    df['last_name'] = df.name.apply(get_last_name)
    
Applying a Lambda to a Row
    row.column_name or row[‘column_name’] to access values of the row
    Use axis = 1 if using apply without specifying column

    # example
    df['Price with Tax'] = df.apply(lambda row:
         row['Price'] * 1.075
         if row['Is taxed?'] == 'Yes'
         else row['Price'],
         axis=1
    )
    
    # example
    total_earned = lambda row: (row.hourly_wage * 40) + ((row.hourly_wage * 1.5) * (row.hours_worked - 40)) \
        if row.hours_worked > 40 \
      else row.hourly_wage * row.hours_worked

    df['total_earned'] = df.apply(total_earned, axis = 1)
```
